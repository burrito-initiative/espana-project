```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Import files
energy_csv = "resources/energy_dataset.csv"
weather_csv = "resources/weather_features.csv"
energy_df = pd.read_csv(energy_csv)
weather_df = pd.read_csv(weather_csv)

# Merge DataFrames into one (big_df)
big_df = pd.merge(energy_df, weather_df, on="time")
```

## Main DataFrame

```python
# Select only necessary columns for a better DataFrame (better_big_df)
better_big_df = pd.DataFrame({
    "Time":big_df["time"],
    "Month":big_df["month"],
    "Day":big_df["day"],
    "Year":big_df["year"],
    "Oil":big_df["generation fossil oil"],
    "Hydro River/Resivoir":big_df["generation hydro run-of-river and poundage"] + big_df["generation hydro water reservoir"],
    "Solar":big_df["generation solar"],
    "Nuclear":big_df["generation nuclear"],
    "City":big_df["city_name"],
    "Temp (F)":round(big_df["temp"]*(9/5)-459.67,2), # converting kelvin to fahrenheit
    "Rain (mm)":big_df["rain_3h"],
    "Snow (mm)":big_df["snow_3h"],
    "Clouds":big_df["clouds_all"],
    "Weather Description":big_df["weather_main"]})

# Preivew summary
better_big_df.count()
```

```python
# Drop any missing data
smaller_better_big_df = better_big_df.dropna(how='any')
smaller_better_big_df.count()
```

### Adding Seasons

```python
seasons = []

# Loop through Month data to assign season name
for month in smaller_better_big_df["Month"]:
    if month <= 2 or month == 12:
        seasons.append("4-Winter")
    if month >= 3 and month <= 5:
        seasons.append("1-Spring")
    if month >= 6 and month <= 8:
        seasons.append("2-Summer")
    if month >= 9 and month <= 11:
        seasons.append("3-Fall")

# Add Seasons to DataFrame
smaller_better_big_df["Season"] = seasons

# Preview
smaller_better_big_df
```

```python
# Energy generation by year by energy type (divided by 5 to remove duplicate city data)
yearly_view = smaller_better_big_df.groupby("Year")["Oil","Hydro River/Resivoir","Solar","Nuclear"].sum()/5

# Energy generation total each year
yearly_view["Year Total"] = yearly_view.sum(axis=1)

yearly_view
```

```python
# Energy generation by type
oil_total = yearly_view["Oil"].sum()
hydro_total = yearly_view["Hydro River/Resivoir"].sum()
solar_total = yearly_view["Solar"].sum()
nuclear_total = yearly_view["Nuclear"].sum()
total_total = yearly_view["Year Total"].sum()
```

```python
# Energy generation total for each energy type
yearly_view_totals = yearly_view
yearly_view_totals.loc['Total'] = yearly_view.sum(axis=0)

yearly_view_totals
```

```python
seasonal_veiw = smaller_better_big_df.groupby("Season")["Oil","Hydro River/Resivoir","Solar","Nuclear"].sum()/5

seasonal_veiw
```

## Madrid
Energy Generation vs Weather

```python
# Create Dataframe for Madrid
madrid_big_df = smaller_better_big_df.loc[better_big_df["City"]=="Madrid"]

# Preview
madrid_big_df.head()
```

```python
yearly_view_madrid = madrid_big_df[["Oil","Hydro River/Resivoir","Solar","Nuclear"]].sum()

yearly_view_madrid
```

```python
energy_sum = yearly_view_madrid.sum()
oil_percent = round(yearly_view_madrid[0]/energy_sum,2)*100
hydro_percent = round(yearly_view_madrid[1]/energy_sum,2)*100
solar_percent = round(yearly_view_madrid[2]/energy_sum,2)*100
nuclear_percent = round(yearly_view_madrid[3]/energy_sum,2)*100

percents = [oil_percent,hydro_percent,solar_percent,nuclear_percent]
labels = ["Oil","Hydro","Solar","Nuclear"]
color = ["red","blue","yellow","green"]
```

```python
plt.pie(percents,labels=labels,colors=color,autopct="%1.1f%%")
plt.show
```

```python
weather_desc = smaller_better_big_df.groupby("Weather Description")["Oil","Hydro River/Resivoir","Solar","Nuclear"].mean()
weather_desc
```

```python
x_limit_desc = np.arange(0,len(weather_desc),1)
plt.bar(x_limit_desc,weather_desc["Nuclear"], color="green")
plt.bar(x_limit_desc,weather_desc["Hydro River/Resivoir"], color="blue")
plt.bar(x_limit_desc,weather_desc["Solar"], color="yellow")
plt.bar(x_limit_desc,weather_desc["Oil"], color="red")
plt.show
```

```python
madrid_weather_timeline = madrid_big_df.groupby(["Year","Month"])["Temp (F)","Clouds"].mean()
madrid_energy_timeline = madrid_big_df.groupby(["Year","Month"])["Oil","Hydro River/Resivoir","Solar","Nuclear"].mean()
madrid_weather_vs_engery_production = pd.merge(madrid_weather_timeline, madrid_energy_timeline, on=["Year","Month"])

madrid_weather_vs_engery_production.head(12)
```

```python
# Bar chart average energy production per month by temp (over 2015-2018)
plt.bar(madrid_weather_vs_engery_production["Temp (F)"],madrid_weather_vs_engery_production["Nuclear"], color="green")
plt.bar(madrid_weather_vs_engery_production["Temp (F)"],madrid_weather_vs_engery_production["Hydro River/Resivoir"])
plt.bar(madrid_weather_vs_engery_production["Temp (F)"],madrid_weather_vs_engery_production["Solar"], color="yellow")
plt.bar(madrid_weather_vs_engery_production["Temp (F)"],madrid_weather_vs_engery_production["Oil"], color="red")

plt.show
```

```python
madrid_seasons = madrid_big_df.groupby(["Year","Season"])["Oil","Hydro River/Resivoir","Solar","Nuclear"].mean()
# madrid_seasons = madrid_seasons[["Spring","Summer","Fall","Winter"]]
madrid_seasons
```

```python
# Energy production by season between 2015-2018
x_season = len(madrid_seasons)
x_limit_madrid = np.arange(0,x_season,1)
plt.bar(x_limit_madrid,madrid_seasons["Nuclear"], color="green")
plt.bar(x_limit_madrid,madrid_seasons["Hydro River/Resivoir"], color="blue")
plt.bar(x_limit_madrid,madrid_seasons["Solar"], color="yellow")
plt.bar(x_limit_madrid,madrid_seasons["Oil"], color="red")

plt.show
```
